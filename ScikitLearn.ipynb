```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart_data=pd.read_csv('heart_failure_clinical_records_dataset.csv')
heart_data
```

# Creating DATA

```python
#Creating INDEPENDENT data.
X=heart_data.drop('DEATH_EVENT', axis=1)
X
```

```python
#Creating DEPENDENT data..
Y=heart_data['DEATH_EVENT']
Y
```

# Choosing ML Model: 

There are multiple ML model but we have to choose as per the requirement. One of the models is RandomForestClassifier. Every model has certain parameters. Performance can be changed using these parameters.

```python

from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.get_params() # showing parameters.
```

# Training/Fitting of Data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3)
```

```python
clf.fit(X_train, Y_train) # Now model is trained although not showing anything.
```

# Evaluation of Model

```python
#X_test data is never seen before by the model. Now, trained_model will do prediction regarding test_data. 
predicted_y=clf.predict(X_test)
predicted_y
```

```python
clf.score(X_train, Y_train) # Testing the trained data.
```

```python
clf.score(X_test, Y_test) #Testing the test data which 88 percent accurate.
```

```python
#Improving Model.
for i in range(10,200,10):
    print(f"Running model with {i} estimator \n")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, Y_train)
    #n_estimator is parameter took from above list shown by .getparamsa
    print(f"Accuracy is {clf.score(X_test, Y_test)} \n")
```

Now, maximum is 88 percent which is found out by putting either 80 or 160 value to i. So, we will give this value i.e 160 or 90 to estimator in order t get directly the best result.

Moreover, model is trained on the basis of some data(X_train, Y_train). Now, trained model is given data( x_test, Y_test) to predict how much this model is accurate. This also implies that how much result of test data is according to train data upon which model is trained.

```python
#Improving Model.
print(f"Running model with {160} estimator \n")
clf = RandomForestClassifier(n_estimators=160).fit(X_train, Y_train)
#n_estimator is parameter took from above list shown by .getparamsa
print(f"Accuracy is {clf.score(X_test, Y_test)} \n")
```
