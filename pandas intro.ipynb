# Reformating

Reformating: if values are given in string type, to convert these given values into integer reformating is used

```python
#before that it was not integer, now it coverted into integer, thats why showing int type. Following mentioned method applied 
Ndata_csv['rme_size_grp']
```

```python
#to replace string character use: .str.replace('[to be replaced]','empty space')
#to covert string type to int type use:astype(int)
Ndata_csv['rme_size_grp'].str.replace('[a,_]','').astype(int)
```

```python
Ndata_missing = pd.read_csv('marine-economy.csv')
```

```python
Ndata_missing
```

```python
#column with empty values.
Ndata_missing['data_value']
```

```python
#Lets show mean values at empty places:NaN. Firstly, we find out the mean.
Ndata_missing['data_value'].mean()
```

```python
#In order to place this mean value in column we use ' Data-name['column'].fillna(Data-name['column'].mean)'
Ndata_missing['data_value'].fillna(Ndata_missing['data_value'].mean())
```

The fed mean value at NaN places is not stored but apparently shown. To store this too, it can also be done using 'inplace=true

```python
Ndata_missing['data_value'].fillna(Ndata_missing['data_value'].mean(), inplace=True)
```

```python
Ndata_missing
```

See the data is stored and change in original data is occured. This task can also be performed by assigning the values to that particular column. In this inplace will not be used.

```python
Ndata_missing['data_value'] = Ndata_missing['data_value'].fillna(Ndata_missing['data_value'].mean())
```

```python
Ndata_missing
```

NaN value can also be removed by using .dropna(). To bring changes in original data .dropna(inplace=true). 

```python
#again to have original data with Nan value we import data file.
Ndata_missing = pd.read_csv('marine-economy.csv')
```

```python
Ndata_missing
```

```python
#Changing in original data can also be performed by assinging this to variable i.e. Ndata_missing = Ndata_missing.dropna() 
Ndata_missing.dropna(inplace = True)
```

```python
Ndata_missing
```

Now we can extract this file too which is cleaned from NaN values. Here the method it

```python
Ndata_missing.to_csv('cleaned-maritime-economy', index=False)
```
